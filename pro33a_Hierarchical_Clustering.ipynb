```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
milk = pd.read_csv(r"C:\Users\admin\Desktop\Datasets\milk.csv",
                  index_col = 0)  # 
#milk = pd.read_csv(r"C:\Users\admin\Desktop\Datasets\\")
#The \ before " will ignore the ", so put\\ 
```

```python
milk.head()
```

```python
scaler = StandardScaler()
```

```python
milk_scaled = scaler.fit_transform(milk)
```

```python
mergs = linkage( milk_scaled, method = "average" )
```

```python
fig_size = plt.rcParams["figure.figsize"]
#fig_size[0] = 12
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size
```

```python
dendrogram(mergs,
           labels = list(milk.index),
           leaf_rotation = 90,
           leaf_font_size= 10,
)
plt.show()
```
