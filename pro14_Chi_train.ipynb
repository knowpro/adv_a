```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
trn = pd.read_csv(r"C:/Hogwarts/advanced_analytics/Datasets/train_v9.csv")
```

```python
trn['Item_Fat_Content'] = trn['Item_Fat_Content'].replace({'low fat': 'Low Fat',
                                                               'LF': 'Low Fat',
                                                              'reg': 'Regular'})
trn
```

```python
ctab = pd.crosstab(trn['Item_Fat_Content'], trn['Item_Type'])
ctab
```

```python
test_statistic, p_value, df, expected_frequencies = chi2_contingency(ctab)
print("p value : ", p_value)
```
