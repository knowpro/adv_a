```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
concr = pd.read_csv(r"C:/Hogwarts/advanced_analytics/Cases/Concrete Strength/Concrete_Data.csv")
```

```python
concr.head()
```

```python
lr = LinearRegression()
```

```python
X = concr.drop('Strength', axis = 1)
y = concr['Strength']
```

```python
poly = PolynomialFeatures( degree = 2 )
X_poly = poly.fit_transform(X)

print(poly.get_feature_names_out())
```

```python
lr.fit(X_poly,y)
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X_poly)
```

```python
print(r2_score(y, y_pred))
```
