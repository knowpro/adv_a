```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
bos = pd.read_csv(r"C:\Users\admin\Desktop\Datasets\boston.csv",
                  index_col = 0)
```

```python
bos.head()
```

```python
scaler = StandardScaler()
bos_scaled = scaler.fit_transform(bos)
```

```python
sil = []

for i in range (2,10):
    km = KMeans(n_clusters = i,
               random_state = 2022)
    km.fit(bos_scaled)
    labels = km.predict(bos_scaled)
    sil.append(silhouette_score(bos_scaled, labels))
```

```python
Ks = np.arange(2,10)
i_max = np.argmax(sil)
print(i_max)
best_k = Ks[i_max]

print("Best k: ", best_k)
# for loop from 2 so refer above working
print(sil[i_max])
```

```python
km = KMeans(n_clusters=4,
           random_state=2022)
km.fit(bos_scaled)
labels = km.predict(bos_scaled)
```

```python
bos["Cluster"] = labels
bos.sort_values("Cluster", inplace=True)
bos
```
