```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
cust = pd.read_csv(r"C:\Users\admin\Desktop\Recency Frequency Monetary\rfm_data_customer.csv",
                  index_col = 0)
```

```python
cust.head()
```

```python
cust = cust.drop(['most_recent_visit'], axis=1)
```

```python
cust.head()
```

```python
scaler = StandardScaler()
```

```python
cust_scaled = scaler.fit_transform(cust)
```

```python
sil = []
for i in range(2,10):
    km = KMeans(n_clusters = i,
               random_state = 2022)
    km.fit(cust_scaled)
    labels = km.predict(cust_scaled)
    sil.append(silhouette_score(cust_scaled,labels))
```

```python
Ks = np.arange(2,10)
i_max = np.argmax(sil)
print(i_max)
best_k = Ks[i_max]

print("Best k: ", best_k)
# for loop from 2 so refer above working
print(sil[i_max])
```

```python
km = KMeans(n_clusters=4,
           random_state=2022)
km.fit(cust_scaled)
labels = km.predict(cust_scaled)
```

```python
cust["Cluster"] = labels
cust.sort_values("Cluster", inplace=True)
cust
```

```python
cust.groupby("Cluster").mean()
```
